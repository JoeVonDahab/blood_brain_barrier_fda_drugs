```python

from B3DB import B3DB_DATA_DICT


df_b3db_reg = B3DB_DATA_DICT["B3DB_classification"]
df_b3db_reg.head()
```

```python
import pandas as pd
fda_drugs = pd.read_csv("fda_drugs_complete.csv")
fda_drugs
```

```python
import re
import pandas as pd

def _normalize(s):
    """Normalize string for exact matching, removing common annotations."""
    if not isinstance(s, str):
        return ""
    s = s.lower()
    s = re.sub(r'\(.*\)', '', s)  # Removes parenthetical content
    s = re.sub(r', .*', '', s)    # MODIFIED: Only remove suffixes after a ', '
    s = re.sub(r'[^a-z0-9\s]', ' ', s)
    s = re.sub(r'\s+', ' ', s).strip()
    return s

def calculate_number_of_drugs_at_each_bbb(fda_drugs=fda_drugs, bbb_data=df_b3db_reg):
    
    # List to store results
    results = []
    
    for index, row in fda_drugs.iterrows():
        drug_name = row["Name"]
        drug_name_norm = _normalize(drug_name)
        
        found_match = False
        
        for idx, bbb_row in bbb_data.iterrows():
            bbb_name = bbb_row["compound_name"]
            bbb_name_norm = _normalize(bbb_name)
            
            # Use strict, exact matching on the normalized names
            if drug_name_norm == bbb_name_norm and drug_name_norm != "":
                
                found_match = True
                bbb_status = bbb_row["BBB+/BBB-"]
                
                print(f"Match found: FDA='{drug_name}' <-> BBB='{bbb_name}' (status={bbb_status})")
                
                # Store matched drug with its BBB status
                results.append({
                    "FDA_Drug_Name": drug_name,
                    "BBB_Matched_Name": bbb_name,
                    "BBB_Status": bbb_status
                })
                break  # Stop after first match
        
        if not found_match:
            # Store unmatched drug with "Unknown" status
            results.append({
                "FDA_Drug_Name": drug_name,
                "BBB_Matched_Name": None,
                "BBB_Status": "Unknown"
            })
    
    # Create DataFrame from results
    results_df = pd.DataFrame(results)
    
    # Print summary statistics
    print("\n" + "="*50)
    print("Summary:")
    print(results_df["BBB_Status"].value_counts())
    
    return results_df

# Run the function
result_df = calculate_number_of_drugs_at_each_bbb()
result_df
```

```python
result_df.to_csv("fda_drugs_bbb_classification.csv", index=False)
```

```python
path_1 = "novel_drugs_small_molecule_bbb_updated (1).xlsx"
small_drugs = pd.read_excel(path_1, sheet_name="Sheet1")
small_drugs
```

```python

def calculate_number_of_drugs_at_each_bbb_small_molecules(fda_drugs=fda_drugs, bbb_data=small_drugs):
    
    num_positive = 0
    num_negative = 0
    num_unknown = 0
    
    for index, row in fda_drugs.iterrows():
        drug_name = row["Name"]
        drug_name_norm = _normalize(drug_name)
        
        found_match = False
        
        for idx, bbb_row in bbb_data.iterrows():
            bbb_name = bbb_row["Drug Name"]
            bbb_name_norm = _normalize(bbb_name)
            
            # Use strict, exact matching on the normalized names
            if drug_name_norm == bbb_name_norm and drug_name_norm != "":
                
                found_match = True
                bbb_status = bbb_row["BBB+/BBB-"]
                
                print(f"Match found: FDA='{drug_name}' <-> BBB='{bbb_name}' (status={bbb_status})")
                
                if bbb_status == "BBB+":
                    num_positive += 1
                elif bbb_status == "BBB-":
                    num_negative += 1
                else:
                    num_unknown += 1 # Should not happen if data is clean
                break  # Stop after first match
        
        if not found_match:
            num_unknown += 1
    
    return {
        "BBB+": num_positive,
        "BBB-": num_negative,
        "Unknown": num_unknown
    }

# Run the function
result = calculate_number_of_drugs_at_each_bbb_small_molecules()
print("\n" + "="*50)
print("Final Results:")
print(result)
```
